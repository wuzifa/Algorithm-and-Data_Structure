```python
#!/usr/bin/env python3

 

# for get ip

import socket, time

import pygame 

from aip import AipSpeech

 

""" 语音技术 APPID AK SK """

SpeechAPP_ID = '120342167'

SpeechAPI_KEY ='L2vnUNNKsnP5EZkDLWhMUf7J'

SpeechSECRET_KEY = '7viPFiN6rAO1qdkOK6dfaPswJlXnu1yf'

 

Speechclient = AipSpeech(SpeechAPP_ID, SpeechAPI_KEY, SpeechSECRET_KEY)

 

#获取IP地址

def getip():

    try:

        s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

        s.connect(('www.baidu.com', 0))

        ip = s.getsockname()[0]

    except:

        ip = "x.x.x.x"

    finally:

        s.close()

    return ip

 

if __name__ == "__main__":

    while(1):

        try:

            ip = getip()

            print(ip)

            if(ip == "x.x.x.x"):
                continue

            if(ip != "x.x.x.x"):

                #result = Speechclient.synthesis(ip, 'zh', 1, options={'spd':3,'vol':9,'per':2,})

                result = Speechclient.synthesis(ip, 'zh', 1, {'spd':3, 'vol' : 1, 'per' : 3} )

 

                #将合成的语音写进文件

                if not isinstance(result,dict):

                    with open('audio.mp3','wb') as f:

                        f.write(result)

                        #我们利用树莓派自带的pygame

                        pygame.mixer.init()

                        pygame.mixer.music.load('./audio.mp3')

                        pygame.mixer.music.play()

                        time.sleep(10)

                        pygame.mixer.init()

                        pygame.mixer.music.load('./audio.mp3')

                        pygame.mixer.music.play()

                        time.sleep(10)

                        break

                else:

                    print(result)

        except Exception:

                raise
```
